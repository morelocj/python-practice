2. Write a function that finds the ship that can carry the most cargo


```python
def most_cargo():
    cargo = []
    c2 = []
    for i in range(1,5):
        response = requests.get(starships + str(i))
        new_response = response.json()
        results = new_response['results']
        for i in results:
            if i['cargo_capacity'].isnumeric():
                i['cargo_capacity'] = int(i['cargo_capacity'])
            cargo.append((i['cargo_capacity'], i['name']))
        for j in cargo:
            if type(j[0]) == int:
                c2.append(j)
        c2.sort() 
    return c2[-1][1]
    
most_cargo()
```

3. Write a function that finds the number of crew and passengers on a given starship

```python
def on_board(ship):
    for i in range(1,5):
        response = requests.get(starships + str(i))
        new_response = response.json()
        results = new_response['results']
        for i in results:
            if i['name'] == ship:
                shipn = i['name']
                crew = i['crew']
                passen = i['passengers']
        return f"There are {crew} crew and {passen} passengers aboard the {shipn}."

on_board('Droid control ship')
```

4. Write a function to find the most expensive starship

```python
def most_exp():
    cost = []
    cost2 = []
    for i in range(1,5):
        response = requests.get(starships + str(i))
        new_response = response.json()
        results = new_response['results']
        for i in results:
            if i['cost_in_credits'].isnumeric():
                i['cost_in_credits'] = int(i['cost_in_credits'])
            cost.append((i['cost_in_credits'], i['name']))
        for j in cost:
            if type(j[0]) == int:
                cost2.append(j)
        cost2.sort() 
    return cost2[-1]
    
most_exp()
```

5. Write a function to find all of the starships that are less than a given price

```python
def less_price(price):
    cost = []
    cost2 = []
    for i in range(1,5):
        response = requests.get(starships + str(i))
        new_response = response.json()
        results = new_response['results']
        for i in results:
            if i['cost_in_credits'].isnumeric():
                if int(i['cost_in_credits']) < price:
                    cost.append((i['cost_in_credits'], i['name']))
    return cost, len(cost)
    
less_price(240000)
```

7. Write a function that finds the shortest and longest ship. Return should be the names of the ship in addition to the numeric values. Convert values into feet.

```python
import re
def short_long():
    length = []
    l2 = []
    for i in range(1,5):
        response = requests.get(starships + str(i))
        new_response = response.json()
        results = new_response['results']
        for i in results:
            if i['length'].isnumeric():
                strip_l = re.sub(',','',i['length'])
                #using regex to get the commas out of a few length values before converting them to float
                value = (float(strip_l)*3.2808)
                #coverting to float and also converting meters to feet
                length.append((value, i['name']))
                #making a list of tuples that pair the length in feet with the name of the ship
    return f"{length[0]} is the shortest and {length[-1]} is the longest."
    
short_long()
```
