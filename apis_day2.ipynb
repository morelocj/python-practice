```python
baseurl = 'https://swapi.dev/api/people/?page='
total_pages = 9

#I need to perform my get request 9 times.
#I want to use a for loop to do that. 
#I can hardcode the values in a for loop, by setting the range with integers
#for i in range(1,10):

data = []

for i in range(1,total_pages+1):
    response = requests.get(baseurl + str(i))
    #'https://swapi.dev/api/people/?page=1'
    #'https://swapi.dev/api/people/?page=2'
    #'https://swapi.dev/api/people/?page=3'
    new_response = response.json()
    data.append(new_response)

pp(data)

```

3. Create a function that returns a list of all the characters that appear in more than x number of films. Where 'x' is the number of films you are checking for and is passed into the function.

```python
def thresh_char(x):
    winners = []
    for i in cleaned_data:
        counter = 0
        films = i['films']
        for film in films:
            counter += 1
        if counter > x:
            winners.append(i['name'])
    return winners

thresh_char(3)
```

4. Create function that returns a tuple of the min, max, and avg height for all charachters

```python
def height_calc():
    char_heights = []
    char_h2 = []
    for i in cleaned_data:
        char_heights.append(i['height'])
        for j in char_heights:
            if j.isnumeric():
                char_h2.append(int(j)) 
    char_h2.sort()
    minim = char_h2[0]
    maxim = char_h2[-1]
    aver = sum(char_h2)//len(char_h2)
    return (minim, maxim, aver)

height_calc()
```

5. Create a function that accepts two arguements: eye color and hair color. Return a dictionary with the min, max, and avg height based on the parameters.

```python
def eyes_hair(eyes,hair):
    char_heights = []
    char_h2 = []
    for i in cleaned_data:
        if i['eye_color'] == eyes and i['hair_color'] == hair:
            char_heights.append(i['height'])
    for j in char_heights:
        if j.isnumeric():
            char_h2.append(int(j)) 
    char_h2.sort()
    results = {}
    results['min_height'] = char_h2[0]
    results['max_height'] = char_h2[-1]
    results['avg_height'] = sum(char_h2)//len(char_h2)
    print(f"For characters with {eyes} eyes and {hair} hair:")
    return results

eyes_hair('brown','brown')
```
